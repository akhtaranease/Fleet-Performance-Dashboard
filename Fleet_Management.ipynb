# Data Cleaning

```python
import pandas as pd;
```

```python
df = pd.read_csv('fFreight.csv');
```

```python
df.head()
```

```python
df['Net Revenue'].head(10)
```

```python
df['Net Revenue'] = df['Net Revenue'].str.replace(',', '.').astype(float)
```

```python
df.head()
```

```python
df['Weight (Kg)'].head(10)
```

```python
df['Weight (Kg)'].head(10)
```

```python
df['Weight (Kg)'] = df['Weight (Kg)'].str.replace(',', '.').astype(float)
```

```python
df['Weight (Kg)'].head(10)
```

```python
df['Weight (Cubic)'].head(10)
```

```python
df['Weight (Cubic)'] = df['Weight (Cubic)'].str.replace(',', '.').astype(float)
```

```python
df['Weight (Cubic)'].head(10)
```

```python
df['Goods Value'].head(10)
```

```python
df['Goods Value'] = df['Goods Value'].str.replace(',', '.').astype(float)
```

```python
df.head()
```

# Data Handling

```python
df['Date'] = pd.to_datetime(df['Date'], format='%Y/%m/%d')
```

```python
df.to_csv("freight_cleaned.csv", index=False)
```
